```python
def get_user_page(user):  
    try:
        driver.get("https://www.instagram.com/" + user + "/")
        time.sleep(5)
        n=driver.find_element(By.XPATH,('.//h2')).text
        p="Sorry, this page isn't available."
        
        if p==n:
            return False
        
        return True
    
    except:
        pass
```

```python
def public():
    try:
        time.sleep(2)
        el = driver.find_element(By.XPATH,"//article")
        
        if len(el.text)!=0:   #private account
            return 0
        
        return 1          #public account

    except:
        pass
    
```

```python
def verified():
    try:
        time.sleep(2)
        m=driver.find_element(By.XPATH,('.//*[contains(text(), "Verified")]')).text
        return 1
    except:
        return 0
    
```

```python
def get_number_of_followers():
    try:
        time.sleep(2)
        followers_count=int(driver.find_element(By.XPATH,('.//*[contains(text(), "followers")]/span')).get_attribute("title").replace(',',''))
        return(followers_count)
    except:
        return -1
```

```python
def get_number_of_following():
    try:
        time.sleep(2)
        total_following=int(driver.find_element(By.XPATH,('.//*[contains(text(), "following")]/span')).text.replace(',',''))
        return(total_following)
    
    except:
        return -1
```

```python
def get_number_of_posts():
    try:
        time.sleep(2)
        posts_count=int(driver.find_element(By.XPATH,('.//*[contains(text(), " posts")]/span')).text.replace(',',''))
        return posts_count
    
    except:
        return -1
```

```python
def get_bio():
    try:
        time.sleep(2)
        el = driver.find_element(By.XPATH,"//ul/following-sibling::div")
#       bio = driver.find_element(By.CLASS_NAME,"_aa_c").text
        x=clean(el.text, no_emoji=True)
        return x

    except:
        return ""
    
```

```python
def extract_primary_prof(biooo):
    try:
        time.sleep(2)
        bio=biooo.split('\n')
        if bio[1].lower() in cats:
            return bio[1]
            
    except:
        return "NaN"
```

```python
def get_bio_links():

    links=[]
    try:
        time.sleep(2)
        el = driver.find_elements(By.XPATH,"//ul/following-sibling::div//a")
        for i in el:
            if "Followed by" not in i.text:
                links.append(i.text)
            
        return links

    except:
        return links
```
